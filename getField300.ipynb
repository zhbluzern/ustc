### Get MARC field 300 a from Alma API for USTC list

This script needs a csv input file with the MMS ids and creates a new csv file with the MMS id and their marc field 300 (subfields a and c). This info is needed for the USTC list and cannot be extracted via Analytics or other means. 

To run this script, add the following:

Write your Alma API key in a file named config.py The file should look like this:

(This file is under .gitignore, the API keys are not publicly shared)

```python
# api_key for your Bib production, read-only

api_key = "API_KEY"
```

This script assumes that the config.py is in the same folder as your script. 
Additionally, adapt the input file name as needed.

```python
import requests
import xml.etree.ElementTree as ET
import csv
import config
import codecs

# Set your Alma API key, base API endpoint, input file name
api_key = config.api_key
base_url = 'https://api-eu.hosted.exlibrisgroup.com/almaws/v1'
# input_file = 'USTC_Export_MMS-IDs-TEST.csv'
input_file = 'USTC_Export_MMS-IDs_ALL.csv'

# Read the input file (use encoding to remove BOM) to get the list of IDs and URLs to update
with codecs.open(input_file, encoding='utf-8-sig') as f:
    id_list = [line.strip().split(',') for line in f]
# debugging: print(id_list)  
# Loop through the list and update each bib record
i = 0
for bib_id, f300a, f300c in id_list:

    #debug: print(id_list[i][0])
    # Construct the API URL to get the bib record in MARCXML format
    bib_url = f'{base_url}/bibs/{bib_id}?view=full&expand=None&apikey={api_key}'

    # Send the request to get the bib record
    response = requests.get(bib_url)

    # Parse the MARCXML data from the response content
    marcxml = ET.fromstring(response.content)

    # Find the 300 field you want to replace
    for field in marcxml.findall('.//datafield[@tag="300"]'):
        # Find the subfield with the URL
        a_subfield = field.find('.//subfield[@code="a"]')
        c_subfield = field.find('.//subfield[@code="c"]')
        if a_subfield is not None:
            # Replace the URL with the new value
            f300a = a_subfield.text
            id_list[i][1] = f300a
            # debugging: print(f300a)
            
        if c_subfield is not None:
            # Replace the URL with the new value
            f300c = c_subfield.text
            id_list[i][2] = f300c
            # debugging: print(f300c)
            
    i+=1
    print("Zeile",i,"verarbeitet:", bib_id, f300a, f300c)
            
# print(id_list)    

# field names 
fields = ['mms_id', '300_a', '300_c'] 
    
# name of csv file 
filename = "ustc_completed.csv"

# writing to csv file 
with open(filename, 'w', newline = '', encoding='utf-8-sig') as csvfile: 
    # creating a csv writer object 
    csvwriter = csv.writer(csvfile, delimiter=';') 
        
    # writing the fields 
    csvwriter.writerow(fields) 
           
    # writing the data rows 
    csvwriter.writerows(id_list)
```
